```python
df_usucap_municip = pd.read_csv("tjma-usucapiao-municipios.csv", sep=";", on_bad_lines="skip")
```

```python
df_usucap_municip
```

```python
df_usucap_municip["Municipio"].value_counts()[lambda x: x > 10]
```

```python
contagens = df_usucap_municip["Municipio"].value_counts()
contagens
```

```python
municipios = contagens[(contagens>10) & (contagens <120)]
municipios
```

```python
municipios_lista = municipios.index.tolist()
municipios_lista
```

```python
df_filtrado = df_usucap_municip[df_usucap_municip["Municipio"].isin(municipios_lista)]
df_filtrado
```

```python
df_filtrado.to_csv("tjma-usucapiao-amostra.csv", index=False, encoding="cp1252", sep=";")
```
